```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 


```

```python
df = pd.read_csv('data/circle_data.csv', delimiter=';')
df.head()

```

```python
df.info()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.axis('equal')
ax.scatter(df['x'], df['y'], c=df['radius'], cmap='seismic', s=df[['radius']]*90)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()


ax.hist(df['radius'], bins=80, color='green', edgecolor='black', linewidth=1.2, density=1)
ax.
```
