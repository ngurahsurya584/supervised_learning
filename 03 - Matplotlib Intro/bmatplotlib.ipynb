```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 10)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.axis('equal')
ax.plot(x, y,)
ax.plot(x, y1)
ax.plot(x, y2)
ax.set(xlabel='x', ylabel='y', title='Plot of y=sqrt(x), y=sin(x), y=cos(x)')
ax.legend(['sqrt(x)', 'sin(x)', 'cos(x)'], loc='upper left', shadow=True)
```
