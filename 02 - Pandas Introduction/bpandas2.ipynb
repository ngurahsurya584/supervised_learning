```python
dfprefer = data[(data.rating>60) | (data.mfr.isin(["K", "N"]))].copy()
dfprefer.head()
```

```python
dfprefer.reset_index(drop=True, inplace=True)
dfprefer.sort_values("rating", ascending=False, inplace=True)
dfprefer
```

```python
data1 = pd.get_dummies(data, columns=["vitamins"]).copy()
data1.head()
```

```python
data1["sodium_bin"]= pd.cut(data1.sodium, bins=[0, 100, 200, data1.sodium.max()], labels=["low", "medium", "high"])
data1.head()
```

```python
data1["rating_bin"]= pd.cut(data1.rating, bins=[0, 40, 70, data1.rating.max()], labels=["C", "B", "A"])
data1.head()
```
